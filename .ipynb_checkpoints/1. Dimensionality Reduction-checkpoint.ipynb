```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

#//numerical integration in python?
```

```python
## Import and clean data
df = pd.read_excel('Material_Database_Prediction.xlsx', skiprows=1, sheet_name='Training and Validation data')
df.columns = df.iloc[0]
df = df[2:]
```

```python
x = df[df.columns[2:17]].dropna() # input data
y = df[df.columns[17:]] # response data
```

```python
## Perform PCA for dimensionality reduction


pca = PCA(n_components=0.90) # use number of components that will explain at least 90% of the data.
pca.fit(x)

print(pca.explained_variance_ratio_)
print(pca.singular_values_)

x_reduced = pca.fit_transform(x)
```

```python
# Convert Principle components to dataframe and join the output data (y). Then save file as CSV.
pd.DataFrame(x_reduced, columns=['PC1', 'PC2', 'PC3']).join(y).to_csv('reduced_data.csv', index=False)
```
